```python
Q1 = df.GarageArea.quantile(0.25)
Q3 = df.GarageArea.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
(df.GarageArea < (Q1 - 1.5 * IQR)) |(df.GarageArea > (Q3 + 1.5 * IQR))
```

```python
df = df.loc[~((df.GarageArea < (Q1 - 1.5 * IQR)) |(df.GarageArea > (Q3 + 1.5 * IQR)))]
```

```python
len(df)
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(df['GarageArea'], df['SalePrice'])
plt.show()
```
